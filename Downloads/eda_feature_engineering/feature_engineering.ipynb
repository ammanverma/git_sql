### Data pre-processing:-

#### Handling categorical features and converting them to numerical :-

```python
## Handling categorical feature "gender"

new['Gender']=new['Gender'].map({'F':0,'M':1})
new.head()
```

```python
## Handling categorical feature "age"

new["Age"].unique()  ## checking how many unique values "age" has.
```

```python
new['Age']=new['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
new.head()
```

```python
## Handling categorical feature "City_Category"

city=pd.get_dummies(new['City_Category'],drop_first=True)
city.head()
```

```python
## drop_first is used because if we have say 3 categories, then 2 are sufficient to represent.
```

```python
## concatinating city with new dataframe

new=pd.concat([new,city],axis=1)
new.head()
```

```python
## dropping city_category feature

df=new.drop('City_Category',axis=1)
df.head()
```

```python
df.isnull().sum()
```

#### Result:- since its a huge missing data, i will replace the missing values.

```python
## exploring the missing values columns

df.Product_Category_2.unique()
```

```python
df.Product_Category_2.value_counts()
```

#### Result:- since only a set of values keep repeating number of times, replacing missing values with MODE will be a better choice.

```python
## replacing 'Product_Category_2' missing values with mode

df['Product_Category_2']=df['Product_Category_2'].fillna(df['Product_Category_2'].mode()[0])
```

```python
## checking the null values after the process

df['Product_Category_2'].isnull().sum()
```

```python
## replacing 'Product_Category_3' missing values with mode

df['Product_Category_3']=df['Product_Category_3'].fillna(df['Product_Category_3'].mode()[0])
```

```python
## checking the null values after the process

df['Product_Category_3'].isnull().sum()
```
