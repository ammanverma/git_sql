### Convert Hours and Minutes into Seconds

```python
def seconds(hours,minutes):
    seconds=(hours*3600)+(minutes*60)
    return(seconds)
hours=int(input("Enter number of hours :"))
minutes=int(input("Enter number of minutes :"))
print("Total seconds are :",seconds(hours,minutes))
```

### Burglary Series (14): Adjectives Total

You call your spouse in anger and a "little" argument takes place.<br> Count the total amount of insults used.<br> Given a dictionary of insults, return the total amount of insults used.<br>

```python
def total_adj(insults):
    total_adj=len(insults)  # finds number of items "len()" in a dictionary
    return(total_adj)
insults={ "a": "moron", "b": "scumbag", "c": "moron", "d": "dirtbag" }
print("Total amount of insults is :",total_adj(insults))
```

### Basketball Points

```python
def Basketball_Points(three_pointers,two_pointers):
    Basketball_Points=three_pointers*3+two_pointers*2
    return(Basketball_Points)
three_pointers=int(input("enter number of 3 pointers :"))
two_pointers=int(input("enter number of 2 pointers :"))
print("Total basket ball points are :", Basketball_Points(three_pointers,two_pointers))
```

### Proper Modulo Operator

```python
def modulous(a,b):
    modulous=a%b       ## gives reminder of two numbers
    return(modulous)
a=float(input("Enter the first number :"))
b=float(input("Enter the second number :"))
print("The reminder is :",modulous(a,b))
```
